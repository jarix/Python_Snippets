# Config Files 101

```python
# !pip install pyyaml

import json
import yaml  # Third-party: pip install pyyaml
import tomllib  # Built-in as of Python 3.11
```

### Read Json

```python
with open('config.json', 'r') as f:
    json_data = json.load(f)
    print(f"JSON Name: {json_data['title']}")
```

### Read YAML

```python
with open('config.yaml', 'r') as f:
    yaml_data = yaml.safe_load(f)
    print(f"YAML Name: {yaml_data['title']}")

print(json_data)
```

### Read TOML

```python
with open('config.toml', 'rb') as f:
    toml_data = tomllib.load(f)
    print(f"TOML Name: {toml_data['title']}")
```

### Check

```python
# All three now exist as standard Python dictionaries
assert json_data['title'] == yaml_data['title'] == toml_data['title']
```
