# Vector Arithmetic 

````python
u = np.array([1,2])
v = np.array([3,4])
print("Vector u: ", u)
print("Vector v: ", v)

# Vector Addition
sumV = u + v
print("Addition: ", sumV)

# Vector Subtraction
subV = u - v
print("Subtraction: ", subV)

# Scalar Multiplication
u2 = u * 2
print("Scalar Multiplication u*2: ", u2)

# Scalar Addition 
u5 = u + 5
print("Scalar Addition (Broadcasting) u+5: ", u5)

# Hadamard product (element-wise product)
prodV = u * v
print("Hadamard (element-wise) product: ", prodV)

# Dot product 
dotV = np.dot(u,v)
print("Dot product: ", dotV)
````

# Matrix Arithmetic

````python
A = np.array([[1,0],[0,1]])
B = np.array([[2,1],[1,2]])
print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix Addition
sumM = A + B
print("Addition:\n", sumM)

# Matrix Subtraction
subM = A - B
print("Subtraction:\n", subM)

# Scalar Multiplication
B2 = B * 2
print("Scalar Multiplication B*2:\n", B2)

# Scalar Addition 
A5 = A + 5
print("Scalar Addition (Broadcasting) A+5:\n", A5)

# Hadamard product (element-wise product)
prodM = A * B
print("Hadamard (element-wise) product:\n", prodM)

# Matrix Multiplication (Dot product)
dotM = np.dot(A,B)
print("Dot product:\n", dotM)

# Matrix Multiplication, alternative Python notattion
multM = A @ B
print("Multiplication:\n", multM)

````

## Reshape Array

````python
# Create a 1D array
array1d = np.array([1, 2, 3, 4, 5, 6])
print(f"Original 1d array: {array1d}")
print(f"Original 1d shape: {array1d.shape}\n")
````

````python
# Reshape to a 2x3 matrix
reshaped_arr_2x3 = array1d.reshape(2, 3)
print(f"Original 1d array: {array1d}")
print(f"Reshaped to 2x3: \n{reshaped_arr_2x3}")
print(f"Shape of 2x3: {reshaped_arr_2x3.shape}\n")
````

When using -1 for one of the dimensions in the new shape, NumPy will automatically calculate the size of that dimension based on the total number of elements and the other specified dimensions.

````python
# Reshape to a 3x2 matrix using -1
reshaped_arr_3x2 = array1d.reshape(3, -1)
print(f"Original 1d array: {array1d}")
print(f"Reshaped to 3x2 (using -1): \n{reshaped_arr_3x2}")
print(f"Shape of 3x2: {reshaped_arr_3x2.shape}\n")
````

````python
# Reshape to a 3x2 matrix using -1
try:
    reshaped_arr = array1d.reshape(-1, -1)
    print(f"Original 1d array: {array1d}")
    print(f"Reshaped array: \n{reshaped_arr}")
    print(f"Shape of reshaped array: {reshaped_arr.shape}\n")
except Exception as e:
    print(f"*** Error: '{e}'")
````

## Squeeze Array

squeeze() function removes single-dimensional entries (axes of length one) from the shape of an array, i.e.,  effectively reduces the number of dimensions in the array without changing the total number of elements.

````python
# Split array by columns with hsplit
array = np.arange(16).reshape(4, 4)
print("Original array:\n", array)
print()

# Split column index 3 to separate array
(array1, array1b) = np.hsplit(array, [1])
print("Split array to 1 column:")
print(array1)
print("Split array1 shape: ", array1.shape)

print()
array2 = np.squeeze(array1)
print("Squeezed array:")
print(array2)
print("Squeezed array2 shape: ", array2.shape)
````

## Stacking Arrays

````python
import numpy as np 

a = np.arange(6)
print("a: ", a)
b = np.arange(start=6, stop=12, step=1)
print("b: ", b)
c = np.arange(start=12, stop=18, step=1)
print("c: ", c)
print()

# Stack back to back horizontally
stack_1 = np.hstack((a, b, c))
print("Horizontal Stack (hstack()):\n", stack_1)
print()

# Stack back to back vertically
stack_2 = np.vstack((a, b, c))
print("Vertical Stack (vstack()):\n", stack_2)
print()

# Stack as rows
stack_r = np.stack((a, b, c), axis=0)  # axis=0 for rows
print("Stack Rows (stack(...,axis=0)):\n", stack_r)
print()

# Stack as columns
stack_c = np.stack((a, b, c), axis=1)  # axis=1 for columns
print("Stack Columns (stack(...,axis=1)):\n", stack_c)
print()
````

## Save to .npy file, Load from .npy file

````python
array = np.array([range(i,i+3) for i in [2,4,6]])

print("Original array:\n", array)

# Note: NumPy automatically appends the .npy extension if you don't provide it
np.save('my_array.npy', array)

# 3. Load the data back from the file
loaded_array = np.load('my_array.npy')

print("Loaded back array:\n", loaded_array)

assert np.array_equal(array, loaded_array), "Arrays are not equal!"

````

## Save multiple arrays to .npz file

````python
# Create sample arrays
array_a = np.array([1, 2, 3])
array_b = np.array([4, 5, 6])

# Save with specific names (recommended)
np.savez('my_archive.npz', features=array_a, labels=array_b)

# Save with compression to reduce file size
np.savez_compressed('my_archive_compressed.npz', features=array_a, labels=array_b)
````

````python
# Load the data
with np.load('my_archive.npz') as data:
    # See all available array names
    print(data.files)  # Output: ['features', 'labels']
    
    # Access specific arrays
    features = data['features']
    labels = data['labels']
    
    print(features) 
    print(labels)

with np.load('my_archive_compressed.npz') as data:
    # See all available array names
    print(data.files)  # Output: ['features', 'labels']
    
    # Access specific arrays
    features = data['features']
    labels = data['labels']
    
    print(features) 
    print(labels)
````
