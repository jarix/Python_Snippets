# Zip() 101

zip() takes two or more iterables (lists, tuples, strings, etc.) and pairs their elements together by position, producing an iterator of tuples:

```python
cities = ["New York", "Seattle", "Helsinki"]
populations = [10.0, 1.0, 0.6]

for city, population in zip(cities, populations):
    print(f"City: {city}, Population: {population} Millions")
```
