## plt.figure() and plt.axes()

plt.Figure() can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels. The axes (an instance of the class plt.Axes) is a bounding box with ticks and labels, which will eventually contain the plot elements that make up the visualization.

```python
fig = plt.figure()
ax = plt.axes()
```

```python
# Add a single plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
plt.show()
```

```python
# Set the Major ticks
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])
# Display Minor ticks
ax.minorticks_on()
# Set the minor ticks for x-axis
ax.set_xticks(np.arange(0, 11, 0.5), minor=True)
# Set the minor ticks for y-axis
ax.set_yticks(np.arange(-1, 1.1, 0.25), minor=True)
fig
```

```python
# Set the Major ticks with custom labels
ax.set_xticklabels(['zero', 'two', 'four', 'six', 'eight', 'ten'])
ax.set_yticklabels(['minus one', 'minus half', 'zero', 'plus half', 'plus one'])
fig
```

```python
# Plotting with a label to reference in the legend, and with markers
ax.plot(x, y, label='Sine curve', marker='o')
# Display the legend on the plot
ax.legend()
fig
```

## 3D Scatter Plot

```python
# Make some 3D Points
def create_box_points(x_size=1.0, y_size=1.0, z_size=1.0, step=0.05, x_offset=0.0, y_offset=0.0, z_offset=0.0):
    """ Create an box shape points array for testing
    """
    points_array = []
    x_range = abs(x_size/2)
    y_range = abs(y_size/2)
    z_range = abs(x_size/2)
    
    for z in [-z_range, z_range]:
        for x in np.arange(-x_range, x_range, step):
            for y in np.arange(-y_range, y_range, step):
                point = [x + x_offset, y + y_offset, z + z_offset]
                points_array.append(point)

    for y in [-y_range, y_range]:
        for x in np.arange(-x_range, x_range, step):
            for z in np.arange(-z_range, z_range, step):
                point = [x + x_offset, y + y_offset, z + z_offset]
                points_array.append(point)

    for x in [-x_range, x_range]:
        for y in np.arange(-y_range, y_range, step):
            for z in np.arange(-z_range, z_range, step):
                point = [x + x_offset, y + y_offset, z + z_offset]
                points_array.append(point)

    return np.array(points_array)

points_array = create_box_points(x_size=2.0, y_size=2.0, z_size=2.0, step=0.2)
print(f"Points Array type: {type(points_array)}")
print(f"Points Array shape: {points_array.shape}")
```

```python
# Plot in 3D Scatter Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points_array[:,0], points_array[:,1], points_array[:,2], marker='.')
plt.show()
```

### Machine Learning Activation Functions

```python

# Define the sigmoid function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Create an array of x values
x = np.linspace(-10, 10, 400)

# Calculate sigmoid and its derivative
y_sigmoid = sigmoid(x)
y_derivative = sigmoid_derivative(x)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, y_sigmoid, label='Sigmoid Function')
plt.plot(x, y_derivative, label='Sigmoid Derivative', linestyle='--')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define tanh function and its derivative
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2


# Calculate tanh and its derivative
y_tanh = tanh(x)
y_tanh_derivative = tanh_derivative(x)

# Plot tanh and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x, y_tanh, label='Tanh Function')
plt.plot(x, y_tanh_derivative, label='Tanh Derivative', linestyle='--')
plt.title('Tanh Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Define ReLU function and its derivative
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)


# Calculate ReLU and its derivative
y_relu = relu(x)
y_relu_derivative = relu_derivative(x)

# Plot ReLU and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x, y_relu, label='ReLU Function')
plt.plot(x, y_relu_derivative, label='ReLU Derivative', linestyle='--')
plt.title('ReLU Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
