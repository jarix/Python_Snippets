# Matplotlib 101

For inspiration, see https://matplotlib.org/3.1.1/gallery/index.html


```python
# pyplot (part of matplotlib) provides handy functions to manage plot elements
from matplotlib import pyplot as plt
import numpy as np
```

## Scatter Plot

```python
X = [1, 2, 3, 4, 5, 6]
Y = [60, 100, 120, 150, 210, 250]

plt.scatter(X,Y)
plt.show()
```

## Connect Scatter Plot with Lines

```python
# Connect Scatter Plot with Lines
plt.scatter(X,Y)
plt.plot(X,Y)
plt.show()
```

### Add Axis Titles

```python
plt.scatter(X,Y)
plt.plot(X,Y)
plt.title("Odometer vs. Time")
plt.xlabel("Time (h)")
plt.ylabel("Odometer (km)")
plt.show()
```

### Use styles

Style reference:

## Using subplots

```python
plt.style.use('default')
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()
#x = np.random.randint(-10,10, size=20)
x = np.arange(-10,10, step=1, dtype=int)
y = x*x + 2*x -5
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

### Plotting multiple lines

```python
fig, ax = plt.subplots()
x = x = np.arange(0,20, step=1, dtype=int)
y = 2*x
plt.plot(x,y)
plt.plot(x,x+2)
plt.show()
```

### Line styles

https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html

```python
fig, ax = plt.subplots()
plt.style.use('seaborn-whitegrid')
x = np.arange(0,20, step=0.1, dtype=float)
y = 2*x

# Default
plt.plot(x,y)

# Specify red color and solid line
plt.plot(x, 3*np.cos(x), color=(1.0, 0.0, 0.0), linestyle='-')

# Specify green color and dashed line
plt.plot(x, 3*np.sin(x), color=(0.0, 1.0, 0.0), linestyle='--')

# Speficy blue color and dotted line
plt.plot(x, 3*np.sin(x)+10, color='b', linestyle=':')

# Speficy grayscale color and dashes + dots line
plt.plot(x, 3*np.cos(x)+10, color='0.77', linestyle='-.')

# Speficy hex color (magenta) and dotted line
plt.plot(x, 3*np.cos(x)+20, color='#FF00FF', linestyle=':')
```

### Set limits to Axis

```python
fig, ax = plt.subplots()
x = np.arange(0,20, step=0.1, dtype=float)
y = 2*x
plt.xlim(-5,15)
plt.ylim(0,20)
plt.plot(x, 3*np.cos(x), color=(1.0, 0.0, 0.0), linestyle='-')
plt.plot(x, 3*np.cos(x)+20, color='#FF00FF', linestyle=':')
```

## Label Plots

```python
plt.plot(x, np.sin(x), label="sin(x)", color='r')
plt.plot(x, np.cos(x), label="cos(x)", color='g')
plt.legend()   # Show labels
plt.xlabel("x")
plt.ylabel("trig(x)")
```

## 3D Scatter Plot

```python
# Make some 3D Points
def create_box_points(x_size=1.0, y_size=1.0, z_size=1.0, step=0.05, x_offset=0.0, y_offset=0.0, z_offset=0.0):
    """ Create an box shape points array for testing
    """
    points_array = []
    x_range = abs(x_size/2)
    y_range = abs(y_size/2)
    z_range = abs(x_size/2)
    
    for z in [-z_range, z_range]:
        for x in np.arange(-x_range, x_range, step):
            for y in np.arange(-y_range, y_range, step):
                point = [x + x_offset, y + y_offset, z + z_offset]
                points_array.append(point)

    for y in [-y_range, y_range]:
        for x in np.arange(-x_range, x_range, step):
            for z in np.arange(-z_range, z_range, step):
                point = [x + x_offset, y + y_offset, z + z_offset]
                points_array.append(point)

    for x in [-x_range, x_range]:
        for y in np.arange(-y_range, y_range, step):
            for z in np.arange(-z_range, z_range, step):
                point = [x + x_offset, y + y_offset, z + z_offset]
                points_array.append(point)

    return np.array(points_array)

points_array = create_box_points(x_size=2.0, y_size=2.0, z_size=2.0, step=0.2)
print(f"Points Array type: {type(points_array)}")
print(f"Points Array shape: {points_array.shape}")
```

```python
# Plot in 3D Scatter Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points_array[:,0], points_array[:,1], points_array[:,2], marker='.')
plt.show()
```
